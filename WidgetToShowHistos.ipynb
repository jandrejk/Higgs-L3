```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import ipywidgets as widgets
from ipywidgets import interactive, interact, fixed, interact_manual
from IPython.display import display, clear_output
```

```python
directory = 'data/'
data_list = []
for filename in os.listdir(directory):
    if filename.endswith(".csv") : 
        data_list.append(filename)
```

```python
data_dict = dict(zip(data_list,data_list))
```

```python
def GetDF (name) :
    return pd.read_csv('data/'+name)
```

```python
def FillHist(df,column,Nbins) :
    plt.hist(df[column],bins=Nbins)
    plt.show()
```

```python
w= widgets.Dropdown(
    options=data_dict,
    description='file:',
)
```

```python
dataFrameObject = interactive(GetDF, name=w)
display(dataFrameObject)
```

```python
df = dataFrameObject.result
```

```python
df.columns.values
```

```python
interact_manual(FillHist,df=fixed(df),column=dict(zip(df.columns.values,df.columns.values)),Nbins=(10,50))
```
